Q4. Polynomial Plotting with Annotations:

Randomly select the coefficients of a 5th degree polynomial. Set the random seed as the last two
digits of your roll number. Plot the polynomial for -5 ≤ x ≤ 5. Annotate the plot to identify the
maxima and minima of the polynomial.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
# Randomly select coefficients for a 5th degree polynomial
np.random.seed(15)
coefficients =np.random.randint(-5,5,6)
print("The coefficients are:", coefficients)

# Define the polynomial function
def polynomial(x, coeffs):
    return sum(c * x**i for i, c in enumerate(coeffs))

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Calculate y values
y_values = polynomial(x_values, coefficients)

# Plot the polynomial
plt.plot(x_values, y_values,color='black', label='5th Degree Polynomial')
#Annotate the plot to identify the maxima and minima of the polynomial.
plt.annotate('Maxima', xy =(5.1,3600), xytext =(2.5, 2000), arrowprops = dict(facecolor ='red', shrink = 0.05),)
plt.annotate('Minima', xy =(-5.09,-9990), xytext =(-4,-7000), arrowprops = dict(facecolor ='green', shrink = 0.05),)
plt.ylabel('f(x)')
plt.title('Random 5th Degree Polynomial')
plt.grid(True)
plt.legend()
plt.show()
```
